## Time slotter (new_slotter(df))

```python
from datetime import timedelta

def not_overlap(df_slot):
    overlaps = 0
    df_slot = df_slot.reset_index()
    for i, row in df_slot.iterrows():
        for j, rowj in df_slot.iloc[i:].iterrows():
            if row.t1 - timedelta(seconds=1) > rowj.t0 and row.module==rowj.module:
                overlaps += 1
                break

    return overlaps
```

```python
def new_slotter(df):
    modules = list(set(df.module))
    res = pd.DataFrame()
    last_time_day = np.datetime64('2022-02-19 23:59:59.990000')

    
    for m in modules:
        tmp = df[df['module']==m].copy()
        tmp = tmp.reset_index()
        tmp = tmp.sort_values(by=['time_value'])

        for i, row in tmp.iterrows():   # righe di un modulo
            job = row.job_id
            module = row.module
            state = row.state
            time0 = row.time_value
            line = row.line

            if state != 'idle':
                try:
                    time1 = tmp.iloc[i+1].time_value
                    rowj_module = tmp.iloc[i+1].module
                    rowj_job_id = tmp.iloc[i+1].job_id
                    rowj_state = tmp.iloc[i+1].state
                    duration = (time1 - time0).total_seconds()
                    slot_line = {'t0': time0, 't1': time1, 'duration': duration, 'line': line, 'job_id':job, 'state':state, 'module':module, 'next_op': f"Module {rowj_module}, Job {str(rowj_job_id)}, State {str(rowj_state)}" }
                    res = res.append(slot_line, ignore_index=True)
                except Exception as e:
                    print(e)
                
            elif state == 'idle':
                duration = (last_time_day - time0).total_seconds()
                slot_line = {'t0': time0, 't1': last_time_day, 'duration': duration, 'line': line, 'job_id':job, 'state':state, 'module':module, 'next_op': f"End"}
                res = res.append(slot_line, ignore_index=True)
    
    return res
```

## Gantt (draw_gantt())

```python
import plotly.express as px
import pandas as pd

def draw_gantt(df, avoid_idle=True):
    if avoid_idle:
        fig = px.timeline(df[df.state!='idle'], x_start="t0", x_end="t1", y="module", color="state", labels='job_id')
        fig.update_yaxes(autorange="reversed") # otherwise tasks are listed from the bottom up
        fig.show()
        return

    fig = px.timeline(df, x_start="t0", x_end="t1", y="module", color="state", labels='job_id')
    fig.update_yaxes(autorange="reversed") # otherwise tasks are listed from the bottom up
    fig.show()
    return
```
